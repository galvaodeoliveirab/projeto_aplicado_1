```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import glob
import matplotlib.pyplot as plt
```

```python
lista_arquivos = glob.glob("Data/*.csv")
```

```python
def processar_arquivos_csv(lista_arquivos):

    dataframes = []

    for arquivo in lista_arquivos:


        df = pd.read_csv(arquivo, sep=';')

        dataframes.append(df)

    df_final = pd.concat(dataframes, ignore_index=True)

    df_final = df_final.reset_index(drop=True)

    return df_final
```

```python
df = processar_arquivos_csv(lista_arquivos)
df.info()
```

```python
df['Data'] = pd.to_datetime(df['Data'], format='%d/%m/%Y', errors='coerce')
df['Hora (UTC)'] = df['Hora (UTC)'].astype('Int64')
df['Temp. Ins. (C)'] = df['Temp. Ins. (C)'].str.replace(',', '.').astype(float)
df['Temp. Max. (C)'] = df['Temp. Max. (C)'].str.replace(',', '.').astype(float)
df['Temp. Min. (C)'] = df['Temp. Min. (C)'].str.replace(',', '.').astype(float)
df['Umi. Ins. (%)'] = df['Umi. Ins. (%)'].str.replace(',', '.').astype(float)
df['Umi. Max. (%)'] = df['Umi. Max. (%)'].str.replace(',', '.').astype(float)
df['Umi. Min. (%)'] = df['Umi. Min. (%)'].str.replace(',', '.').astype(float)
df['Pto Orvalho Ins. (C)'] = df['Pto Orvalho Ins. (C)'].str.replace(',', '.').astype(float)
df['Pto Orvalho Max. (C)'] = df['Pto Orvalho Max. (C)'].str.replace(',', '.').astype(float)
df['Pto Orvalho Min. (C)'] = df['Pto Orvalho Min. (C)'].str.replace(',', '.').astype(float)
df['Pressao Ins. (hPa)'] = df['Pressao Ins. (hPa)'].str.replace(',', '.').astype(float)
df['Pressao Max. (hPa)'] = df['Pressao Max. (hPa)'].str.replace(',', '.').astype(float)
df['Pressao Min. (hPa)'] = df['Pressao Min. (hPa)'].str.replace(',', '.').astype(float)
df['Vel. Vento (m/s)'] = df['Vel. Vento (m/s)'].str.replace(',', '.').astype(float)
df['Dir. Vento (m/s)'] = df['Dir. Vento (m/s)'].str.replace(',', '.').astype(float)
df['Raj. Vento (m/s)'] = df['Raj. Vento (m/s)'].str.replace(',', '.').astype(float)
df['Radiacao (KJ/m²)'] = df['Radiacao (KJ/m²)'].str.replace(',', '.').astype(float)
df['Chuva (mm)'] = df['Chuva (mm)'].str.replace(',', '.').astype(float)
df.info()
```

```python
df_agrupado1 = round(df.groupby('Data', as_index=False)['Temp. Ins. (C)'].mean(),1)
df_agrupado2 = round(df.groupby('Data', as_index=False)['Temp. Max. (C)'].mean(),1)
df_agrupado3 = round(df.groupby('Data', as_index=False)['Temp. Min. (C)'].mean(),1)
df_agrupado4 = round(df.groupby('Data', as_index=False)['Umi. Ins. (%)'].mean(),1)
df_agrupado5 = round(df.groupby('Data', as_index=False)['Umi. Max. (%)'].mean(),1)
df_agrupado6 = round(df.groupby('Data', as_index=False)['Umi. Min. (%)'].mean(),1)
df_agrupado7 = round(df.groupby('Data', as_index=False)['Pto Orvalho Ins. (C)'].mean(),1)
df_agrupado8 = round(df.groupby('Data', as_index=False)['Pto Orvalho Max. (C)'].mean(),1)
df_agrupado9 = round(df.groupby('Data', as_index=False)['Pto Orvalho Min. (C)'].mean(),1)
df_agrupado10 = round(df.groupby('Data', as_index=False)['Pressao Ins. (hPa)'].mean(),1)
df_agrupado11 = round(df.groupby('Data', as_index=False)['Pressao Max. (hPa)'].mean(),1)
df_agrupado12 = round(df.groupby('Data', as_index=False)['Pressao Min. (hPa)'].mean(),1)
df_agrupado13 = round(df.groupby('Data', as_index=False)['Vel. Vento (m/s)'].mean(),1)
df_agrupado14 = round(df.groupby('Data', as_index=False)['Dir. Vento (m/s)'].mean(),1)
df_agrupado15 = round(df.groupby('Data', as_index=False)['Raj. Vento (m/s)'].mean(),1)
df_agrupado16 = round(df.groupby('Data', as_index=False)['Radiacao (KJ/m²)'].mean(),1)
df_agrupado17 = round(df.groupby('Data', as_index=False)['Chuva (mm)'].mean(),1)
```

```python
df_index = df['Data'].drop_duplicates().reset_index(drop=True)
```

```python
df_index = pd.merge(df_index, df_agrupado1, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado2, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado3, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado4, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado5, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado6, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado7, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado8, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado9, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado10, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado11, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado12, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado13, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado14, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado15, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado16, on='Data', how='left')
df_index = pd.merge(df_index, df_agrupado17, on='Data', how='left')
df_index
```

```python
df_ordenado = df_index.sort_values(by='Data', ascending=True)
```

```python
def calcular_estatisticas(df):
  cols_float = df.select_dtypes('float64')

  estatisticas = {
      'media': round(cols_float.mean(),1),
      'mediana': round(cols_float.median(),1),
      'desvio_padrao': round(cols_float.std(),1),
      'primeiro_quartil': round(cols_float.quantile(0.25),1),
      'terceiro_quartil': round(cols_float.quantile(0.75),)
  }

  return estatisticas
```

```python
estatisticas = calcular_estatisticas(df_ordenado)
```

```python
for estatistica, valores in estatisticas.items():
  print(f"{estatistica}:\n{valores}\n")
```

```python
df_ordenado.set_index('Data', inplace=True)
```

```python
# Resample para agrupar por períodos de 30 dias e calcular a média
df_resampled = df_ordenado.resample('30D').mean()
```

```python
# Configurando o gráfico
plt.figure(figsize=(10, 6))

# Plotando a média da variável 'Temp. Min. (C)'
plt.plot(df_resampled.index, df_resampled['Temp. Min. (C)'], marker='o', color='blue', label='Média de Temp Min')

# Personalizando o gráfico
plt.title('Média da Temperatura Mínima a Cada 30 Dias')
plt.xlabel('Data')
plt.ylabel('Temperatura Mínima (°C)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# Configurando o gráfico
plt.figure(figsize=(10, 6))

# Plotando a média da variável 'Temp. Max. (C)'
plt.plot(df_resampled.index, df_resampled['Temp. Max. (C)'], marker='o', color='red', label='Média de Temp Max')

# Personalizando o gráfico
plt.title('Média da Temperatura Máxima a Cada 30 Dias')
plt.xlabel('Data')
plt.ylabel('Temperatura Máxima (°C)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```
